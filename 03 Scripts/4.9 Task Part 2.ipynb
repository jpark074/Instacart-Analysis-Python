# Step 5

```python
# Check whether there's a difference in expenditure by creating line graph
# Create sample with 70/30 split

np.random.seed(4)
```

```python
# Create a list holding True/False values to test np.random.rand() <= 0.7

dev = np.random.rand(len(df_ords_prods_cust)) <= 0.7
```

```python
# Store 70% of the sample in the df big

big = df_ords_prods_cust[dev]
```

```python
# Store 30% of the sample in the df small

small = df_ords_prods_cust[~dev]
```

```python
len(df_ords_prods_cust)
```

```python
# Check results

len(big) + len(small)
```

```python
# Speed up processing by reducing sample to only "order_hour_of_day" & "prices"

df_line = small[['order_hour_of_day', 'prices']]
```

```python
# Create line graph of "prices" by order_hour_of_day"
line_hour_prices = sns.lineplot(data = df_line, x = 'order_hour_of_day', y = 'prices')
line_hour_prices.locator_params(nbins=24)
```

# Step 6

```python
# Create line chart of connection between "age" and "n_dependents"

df_line_2 = small[['age', 'n_dependants']]
```

```python
# Create line graph of "n_dependants" by "age"

line_age_dependants = sns.lineplot(data = df_line_2, x = 'age', y = 'n_dependants')
```

Based on the line graph, there does not seem to be a connection between age and number of dependents as line 
fluctuates extremely with no clear pattern. Some interesting things to note, however, is that around the age 60
had the most dependents with age 66-67 and 18-19 following behind. Furthermore, the only discernable trend is
around the age of 40, where the line chart still fluctuates but increases overall.

# Step 7

```python
# Create scatterplot between "age" and "income"

scatterplot_age_income = sns.scatterplot(x = 'age', y = 'income', data = df_ords_prods_cust)
```

The scatterplot shows that the vast majority of customers at all ages tend to make under 200k in income. 
This is shown by the majority of data points clustering around 200k and below. Additionally, no one under
the age of 40 seems to be making above 400k. Overall, it seems that income tends to generally increase around
the age of 40.
