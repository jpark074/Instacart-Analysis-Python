# Aggregating Data with agg()

```python
# Split the data into groups & apply agg()

df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Using the mean function

df_agg = df.groupby('department_id')['order_number'].mean()
```

# Aggregating Data with transform()

```python
# Aggregate the data and apply the transform function to create new column

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check basic statistics of the product prices for each loyalty category

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

The price of products purchased by loyal customers does differ from new and regular customers. The average price of the products purchased by loyal customers is lower than that of new and regular customers, with new customers having the highest average purchase price. 

The maximum price is 99999 which is most likely and error and would have skewed the distribution. Further analysis is needed.

```python
# Check if there any other prices that seem to be errors

ords_prods_merge['prices'].sort_values(ascending = False).unique()
```

```python
# Fix prices by copying prices column into new column fixed_prices and changing 99999 to 9.99 and 14900 to 1.49

ords_prods_merge['fixed_prices'] = ords_prods_merge['prices']
```

```python
ords_prods_merge = ords_prods_merge.replace({"fixed_prices": {99999.0 : 9.9, 14900.0 : 1.49}})
```

```python
# Check 

ords_prods_merge['fixed_prices'].max()
```

With the prices fixed we can see that there is a very small difference in the average price of the products purchased by customers.

# Step 6

```python
# Create average_price column

ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['fixed_prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
# Create spending flag based on average_price across all their orders

ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag' ] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.head()
```

```python
# Check the frequency

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# Step 7 

```python
# Create median_last_order column

ords_prods_merge['median_last_order'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merge.head(15)
```

```python
# Create last_order_frequency flag

ords_prods_merge.loc[ords_prods_merge['median_last_order'] > 20, 'last_order_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_last_order'] > 10) & (ords_prods_merge['median_last_order'] <= 20), 'last_order_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_last_order'] <= 10, 'last_order_frequency'] = 'Frequent customer'
```

```python
ords_prods_merge.head()
```

```python
# Check frequency

ords_prods_merge['last_order_frequency'].value_counts(dropna = False)
```
